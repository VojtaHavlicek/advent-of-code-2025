```python
from collections import deque

with open("input.txt", "r") as file:
    data = file.read()
    data = data.strip().split('\n\n')

    shapes = data[:-1]
    shapes_dict = {}
    for shape in shapes:
        entry = shape.split(':')
        label = int(entry[0])
        shapes_dict[label] = entry[1].strip().split('\n')
    shapes = shapes_dict.values()

    regions_list = []
    regions = data[-1].split('\n')
    for region in regions:
        row = region.split(':')
        label = row[0]
        counts = list(map(int, row[1].split()))
        regions_list.append((label, counts))
    regions = regions_list

def rotate(shape):
    rows = len(shape)
    cols = len(shape[0])
    new_shape = [['.' for _ in range(rows)] for _ in range(cols)]
    for r in range(rows):
        for c in range(cols):
            new_shape[c][rows - 1 - r] = shape[r][c]
    return [''.join(row) for row in new_shape]

def flip(shape):
    return [row[::-1] for row in shape]

def all_transformations(shape):
    transformations = set()
    current_shape = shape
    for _ in range(4):
        current_shape = rotate(current_shape)
        transformations.add(tuple(current_shape))
        transformations.add(tuple(flip(current_shape)))
    return [list(t) for t in transformations]

def can_place(board, shape, r, c):
    """ Check if shape can be placed on board at position (r, c) """
    for sr, srow in enumerate(shape):
        for sc, scell in enumerate(srow):
            if scell == '#':
                br = r + sr
                bc = c + sc
                if br >= len(board) or bc >= len(board[0]) or board[br][bc] == '#':
                    return False
    return True


print(len(regions))
overflow_boxes = 0
no_interlock = 0

#print(shapes)

for region in regions:
    label, counts = region

    # ---> Initialize board
    ROWS, COLS = tuple(int(c) for c in label.split('x'))
    board = [['.' for _ in range(ROWS)] for _ in range(COLS)]

    # ---> The total area of boxes has to be smaller than the board area: skip if not.
    areas = [sum(row.count('#') for row in shape) for shape in shapes]
    transforms = [all_transformations(shape) for shape in shapes]

    total_boxes = sum([counts[idx]*areas[idx] for idx in range(len(shapes))])
    board_area = ROWS * COLS
    if total_boxes > board_area:
        #print(f"Region {label} cannot be filled: required area {total_boxes}, board area {board_area}")
        overflow_boxes += 1
        continue

    # ---> Try placing boxes without interlock. All blocks are 3x3, same size
    if (COLS//3) * (ROWS//3) >= sum(counts):
        #print(f"Region {label} can be filled without interlock.")
        no_interlock += 1
        continue



print(f"Total regions that cannot be filled: {overflow_boxes} | can be filled without interlock: {no_interlock}")
# LOL 
```
